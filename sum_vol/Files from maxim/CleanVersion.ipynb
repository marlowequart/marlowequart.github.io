```python
import pandas as pd
import numpy as np
```

Create sample dataframe and reindex it

```python
df1 = pd.DataFrame({'Date':[20160501,20160501,20160501,20160501,20160502,20160502],
                     'Time':['0900','0905','0913','0915','0920','0925'],
                 'Volume':[1,2,3,4,5,6],
                 'Open':[10,35,20,15,5,55],
                 'High':[10,35,20,15,5,55],
                 'Low':[10,35,20,15,5,55],
                 'Close':[10,20,15,20,0,45]}
                ,columns=['Date', 'Time','Volume', 'Open', 'High', 'Low', 'Close'])
df1
```

```python
df1 = df1.set_index(['Date', 'Time'])
df1
```

Create sample dataframe and reindex it

```python
df2 = pd.DataFrame({'Date':[20160501,20160501,20160501,20160501,20160502,20160602],
                     'Time':['0900','0902','0910','0915','0920','0900'],
                 'Volume':[1,2,3,4,5,6],
                 'Open':[10,35,20,15,5,55],
                 'High':[10,35,20,15,5,55],
                 'Low':[10,35,20,15,5,55],
                 'Close':[10,20,15,20,0,45]}
                ,columns=['Date', 'Time','Volume', 'Open', 'High', 'Low', 'Close'])
df2
```

```python
df2 = df2.set_index(['Date', 'Time'])
df2
```

Now merge two dataframes and fill NA's (missed data) with 0

```python
df3 = pd.merge(df1, df2, how='outer', on=['Date', 'Time'])
df3.fillna(0,inplace=True)
df3
```

```python
df3['Volume']=df3['Volume_x']+df3['Volume_y']
df3
```

```python
o1=df3['Open_x']
o2=df3['Open_y']
h1=df3['High_x']
h2=df3['High_y']
l1=df3['Low_x']
l2=df3['Low_y']
c1=df3['Close_x']
c2=df3['Close_y']
```

Now just to be sure that everything is ok we look at the indexes of the empty records in first dataframe (o1 as a part of this dataframe).
After that we will fill this skipped records from second dataframe (merged columns with postfix _y)

```python
for i in filter(lambda x: o1[x] == 0, range(len(o1))):
    print('Index, skipped in o1: {}'.format(i))
```

```python
for x in range(len(o1)):
    for i in filter(lambda x: o1[x] == 0, range(len(o1))):
        o1[i] = o2[i]
        h1[i] = h2[i]
        l1[i] = l2[i]
        c1[i] = c2[i]
df3["Open"]=o1
df3["High"]=h1
df3["Low"]=l1
df3["Close"]=c1
df3 = df3.drop(columns=['Open_x', 'High_x', 'Low_x', 'Close_x', 'Volume_x',
                'Open_y', 'High_y', 'Low_y', 'Close_y', 'Volume_y'])
df3
```

```python
# Now just sort it in desired order
df3 = df3.sort_values(by=['Date','Time'], ascending=True)
df3
```
